```python
# Downloading Data From https://www.kaggle.com/datasets/allen-institute-for-ai/CORD-19-research-challenge

import os 
import subprocess
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent # Getting Parent path 
DATA_DIR = PROJECT_ROOT/'data'/'raw'
DATA_NAME = 'josephassaker/covid19-global-dataset'


DATA_DIR.mkdir(parents=True, exist_ok=True)

# Downloading dataset 
print(' ------- Downloading Dataset -------')
print(f'Downloading {DATA_NAME}')
print(f'Data saving to {DATA_DIR}')

subprocess.run([
    'kaggle',
    'datasets',
    'download',
    '-d',
    DATA_NAME,
    '-p',
    str(DATA_DIR),
    '--unzip'
],
    check=True)

print(' ------- Download complete -------')
```
